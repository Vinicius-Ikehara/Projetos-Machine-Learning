```python
#Ao invés de NaN vamos trabalhar com 0 no lugar, já que as colunas restantes trabalham com int ou float
df1.fillna(value = 0, inplace = True) 
```

```python
pd.set_option('display.max_columns', None)
df1.head()
```

```python
X = df1.iloc[:,1 :-1].values    # Nossa Característica => Anos de Experiência (Variável Independente) 
y = df1.iloc[:, -1].values 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression() # Modelo de Regressão básico do Scikit-Learn
model.fit(X_train, y_train)
```

```python
# O modelo agora vai pegar os valores dos anos de trabalho do X test
# E vai utiliza-los para tentar prever qual será o seu salário, já que o modelo já foi treinado para isso
predicao = model.predict(X_test)
```

```python
erro = abs(predicao - y_test)
resultados = pd.DataFrame({'Valor da Predição':np.round(predicao), 'Valor Real':y_test, 'Erro':np.round(erro)})
resultados.head(15)
```
